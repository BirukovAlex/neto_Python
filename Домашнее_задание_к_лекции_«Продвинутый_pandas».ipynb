**Задание 1**
Для датафрейма log из материалов занятия создайте столбец source_type по правилам:

если источник traffic_source равен Yandex или Google, то в source_type ставится organic;

для источников paid и email из России ставим ad;

для источников paid и email не из России ставим other;

все остальные варианты берём из traffic_source без изменений.

```python
log = pd.read_csv('/content/visit_log.csv', sep=';')
log.head()

#условие 1 Yandex или Google, то в source_type ставится organic
log.loc[
    (log.traffic_source  == 'yandex') |
     (log.traffic_source  == 'google'),
    'source_type'] = 'organic'

#условие 2 для источников paid и email из России ставим ad;
log.loc[
    (
        (log.traffic_source  == 'paid') |
        (log.traffic_source  == 'email')
        ) &
    (log.region == 'Russia'),
    'source_type'] = 'ad'

#условие 3 для источников paid и email не из России ставим other;
log.loc[
    (
        (log.traffic_source  == 'paid') |
        (log.traffic_source  == 'email')
        ) &
    (log.region != 'Russia'),
    'source_type'] = 'other'

#условие 4 все остальные варианты берём из traffic_source без изменений;
log.loc[log.source_type == 'nan', 'source_type'] = log['traffic_source']

log.head()
```

**Задание 3**
Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значениями столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('/content/ratings.csv')
groupby_userId = ratings.groupby('userId').agg({'timestamp':['count','min','max']})['timestamp'].reset_index()
top100 = groupby_userId.loc[(groupby_userId['count'] > 100)]  #фильтруем больше 100
top100_lifetime = (top100['max'] - top100['min']).mean() #считаем среднюю продолжительность
top100_lifetime
```

**Задание 4**
Дана статистика услуг перевозок клиентов компании по типам:

1. rzd - железнодорожные перевозки
2. auto - автомобильные перевозки
3. air - воздушные перевозки
4. client_base - адреса клиентов

Нужно сформировать две таблицы:

1. таблицу с тремя типами выручки для каждого client_id без указания адреса клиента;
2. аналогичную таблицу по типам выручки с указанием адреса клиента.
*3. Обратите внимание, что в процессе объединения таблиц данные не должны теряться.*

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1',
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# джойним перевозки
all_transfers_no_address = rzd.merge(auto, how='left', on='client_id').\
                               merge(air, how='left', on='client_id')
all_transfers_no_address
```

```python
# к перевозкам джойним клиентскую базу
#вариант 1. джойним к полученной таблице базу клиентов (справа)
all_transfers_address_1 = all_transfers_no_address.merge(client_base, how='right', on='client_id')
```

```python
all_transfers_address_1
```

```python
#вариант 2. сразу джойним всё к клиентской базе (слева)
all_transfers_address_2 = client_base.merge(rzd, how='left', on='client_id').\
                            merge(auto, how='left', on='client_id').\
                            merge(air, how='left', on='client_id')
```
