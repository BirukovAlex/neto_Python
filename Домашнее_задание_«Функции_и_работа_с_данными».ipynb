**Задание 1**


> Напишите функцию, которая классифицирует фильмы из материалов занятия по правилам:
>
> оценка 2 и ниже — низкий рейтинг;
>
> оценка 4 и ниже — средний рейтинг;
>
> оценка 4.5 и 5 — высокий рейтинг.
>
> Результат классификации запишите в столбец class.

```python
import pandas as pd
```

```python
#загружаем и джойним две таблицы, чтобы было видно названия фильмов
movies = pd.read_csv('/content/movies.csv')
ratings = pd.read_csv('/content/ratings.csv')
mov_rat = ratings.merge(movies, on = 'movieId')
mov_rat.head()
```

```python
#пишем функцию перебора значений по столбцу rating
def movie_class(row):
  if row ['rating'] <= 2.0:
    return 'Низкий рейтинг'
  elif 2.0 < row ['rating'] <= 4.0:
    return 'Средний рейтинг'
  else:
    return 'Высокий рейтинг'
```

```python
#применяем функцию к объединенной таблицe mov_rat и записываем класс в новый столбец class
mov_rat['class'] = mov_rat.apply(movie_class, axis=1)
mov_rat.head()
```

**Задание 2**
Используйте файл keywords.csv.

Нужно написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определённому региону.

Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.


Правила распределения по регионам Центр, Северо-Запад и Дальний Восток:

geo_data = {
'Центр': ['москва', 'тула', 'ярославль'],

'Северо-Запад': ['петербург', 'псков', 'мурманск'],

'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']

}

Результат классификации запишите в отдельный столбец region.

```python
# загружаем дата сет с ключевыми словами
keywords = pd.read_csv('/content/keywords.csv')
```

```python
# каждый запрос превращаем в список слов
def searching_words(row):
  word = row['keyword'].split(' ')
  return word
```

```python
#составляем список городов
geo_data = {
'Центр': ['москва', 'ютуб','тула', 'ярославль', 'ялте', 'оскол'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
}
city = []
for values in geo_data.values():
  for i in values:
    city.append(i)
city
```

```python
#добавляем столбец со списокм слов из запроса, чтобы потом в нём искать город
keywords['splited'] = keywords.apply(searching_words, axis=1)
keywords.head()
```

```python
# функция, которая ищет город из списка в списке слов запроса
def find_city(row):
  i=0
  for items in city:
    if items in row['splited']:
      i+=1
      return items
  if i==0:
      return 'empty'
# сразу добавляем город в отдельный столбец
keywords['city'] = keywords.apply(find_city, axis=1)
keywords.head(200)
```

```python
# проверяем работает ли цикл по городу
keywords[(keywords['city'] != 'empty') & (keywords['city'] != 'москва')]
```

```python
#добавляем функцию, которая сопоставит город из столбца город и значения словаря geo_data, если найдёт - поставит значение ключа в таблицу
def region(row):
  i=0
  for items in geo_data.items():
      if row['city'] in items[1]:
        i+=1
        return items [0]
  if i ==0:
        return 'undefined'

keywords['region'] = keywords.apply(region, axis=1)
keywords.head()
```

```python
#проверяем работает ли цикл по региону
keywords[(keywords['region'] == 'Северо-Запад')]
```
