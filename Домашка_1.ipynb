Задание 1 Даны 2 переменных, в которых хранятся строки произвольной длины: phrase_1 и phrase_2. Напишите код, который проверяет какая из этих строк длиннее.

```python
# консоль с вводом текста, тип данных str
phr_1 = input('Введите фразу 1: ')
phr_2 = input('Введите фразу 2: ')

# на всякий случай запишем длины фраз в отдельные переменные
p1 = len(phr_1)
p2 = len(phr_2)
#print(p1)
#print(p2)

#операция сравнения
if p1 > p2:
  print('Фраза 1 длиннее Фразы 2')
elif p1 < p2:
  print('Фраза 2 длиннее Фразы 1')
else:
  print('Фразы равны')
```

Задание 2 Дана переменная, в которой хранится число (год). Необходимо написать программу, которая выведет, является ли данный год високосным или обычным.

Пример работы программы: 1.

year = 2020 Результат: Високосный год

year = 2019 Результат: Обычный год

```python
# консоль с вводом года

year = int(input('Введите год: '))

if(year % 4 == 0 and year % 100 != 0) or year %400 == 0:
  print('Високосный год')
else:
  print('Обычный год')
```

```python
day = int(input('Введите день: '))
month = input('Введите месяц: ').lower() #приведение к строчным символам, чтобы интерпретатор воспринимал разные буквы, напечатанные пользователем (Декабрь, ДЕКАБРЬ и т.д.)

if(month == 'март' and day >= 21) or (month == 'апрель' and day <= 19):
    reslt = 'Овен'
elif(month == 'апрель' and day >= 20) or (month == 'май' and day <= 20):
    reslt = 'Телец'
elif(month == 'май' and day >= 21) or (month == 'июнь' and day <= 21):
    reslt = 'Близнецы'
elif(month == 'июнь' and day >= 22) or (month == 'июль' and day <= 22):
    reslt = 'Рак'
elif(month == 'июль' and day >= 23) or (month == 'август' and day <= 22):
    reslt = 'Лев'
elif(month == 'август' and day >= 23) or (month == 'сентябрь' and day <= 22):
    reslt = 'Дева'
elif(month == 'сентябрь' and day >= 23) or (month == 'октябрь' and day <= 23):
    reslt = 'Весы'
elif(month == 'октябрь' and day >= 24) or (month == 'ноябрь' and day <= 22):
    reslt = 'Скорпион'
elif(month == 'ноябрь' and day >= 23) or (month == 'декабрь' and day <= 21):
    reslt = 'Стрелец'
elif(month == 'декабрь' and day >= 22) or (month == 'январь' and day <= 19):
    reslt = 'Козерог'
elif(month == 'январь' and day >= 20) or (month == 'февраль' and day <= 18):
    reslt = 'Водолей'
elif(month == 'февраль' and day >= 19) or (month == 'март' and day <= 20):
    reslt = 'Рыбы'
else:
    reslt = 'Проверьте данные, пожалуйста'

print(f'Ваш знак зодиака: {reslt}')
```

Задание 4 Вам нужно написать программу для подбора упаковок по размерам товара. Размеры (ширина, длина, высота) хранятся в переменных (в сантиметрах):

Используйте следующие правила:

если каждое из трех измерений менее или равно 15 сантиметрам, то выведите на экран 'Коробка №1'; если хотя бы одно из измерений больше 2 метров, то выводите 'Упаковка для лыж'; если хотя бы одно из измерений больше 15 сантиметров, но менее 50 сантиметров, то выводите 'Коробка №2'; во всех остальных случаях выводите 'Коробка №3'. Пример работы программы: 1.

width = 15 length = 55 height = 15 Результат: Коробка №3

width = 45 length = 205 height = 45 Результат: Упаковка для лыж

```python
width = int(input('Ширина: '))
length = int(input('Длина: '))
height = int(input('Высота: '))

if width <= 15 and length <= 15 and height <= 15:
  print('Коробка №1')
elif(15 < width < 50 or 15 < length < 50 or 15 < height < 50):
  print ('Коробка №2')
elif(width > 200 or length > 200 or height > 200):
  print('Упаковка для лыж')
else:
  print('Коробка №3')
```

Задание 5 (необязательное) Дана переменная, в которой хранится шестизначное число (номер проездного билета). Напишите программу, которая будет определять, является ли данный билет 'счастливым'. Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера.

Примеры работы программы:

number = 123456 Результат: Несчастливый билет

number = 123321 Результат: Счастливый билет

```python
#логика: консоль с вводом номера, сумма трех первых цифр, сумма трех последних цифр, сравнение двух сумм
nmb = list(map(int,input('Введите номер билета: ')))
summa_first = nmb[0]+nmb[1]+nmb[2]
summa_last = nmb[3]+nmb[4]+nmb[5]
"""print(summa_first)
print(summa_last)"""

if summa_first == summa_last:
  print('Счастливый билет')
else:
  print('Несчастливый билет')
```
